```python
import pandas as pd
import matplotlib as plt
```

```python
glims = pd.read_csv('glims_csvutf8.csv')
glims.head(20)
#can see that there are repeated rows in dataframe
```

```python
#Want to isolate unique ANLYS_ID
#variables we care about: ANLYS_ID, GLAC_ID, AREA, DB_AREA, WIDTH, LENGTH, PRIMECLASS, MIN_ELEV, MEAN_ELEV, MAX_ELEV, 
# SRC_DATE, 'REC_STATUS', 'GLAC_NAME', 'GLAC_STAT', 'GEOG_AREA', 'PARENT_ID'
glims_unique = glims.drop_duplicates(subset=['ANLYS_ID'])
glims_clean = glims_unique.drop(labels=['LINE_TYPE','ANLYS_TIME','WGMS_ID','LOCAL_ID','SUBM_ID','RELEASE_DT'], axis=1)
glims_clean = glims_clean.drop(labels=['PROC_DESC','RC_ID','CHIEF_AFFL','SUBMITTERS','ANALYSTS'], axis=1)
glims_clean.head(10)
```

```python
#df1['e'] = Series(np.random.randn(sLength), index=df1.index)
glims_clean['date'] = pd.to_datetime(glims_clean['SRC_DATE'])
glims_clean['year'] = glims_clean['date'].dt.year
glims_clean['month'] = glims_clean['date'].dt.month
```
