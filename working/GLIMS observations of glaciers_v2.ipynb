```python
#Let's try to get the "good IDs"

#To start, these are the glaciers with more than 3 observations
allIDs = glims.GLAC_ID.unique()
goodIDs = []
for i in allIDs:
    if numGlacierObs[i] > 3:
        goodIDs.append(i)
    
print(len(goodIDs))


```

```python
greatIDs = []

#Now find glaciers with 4 or more obs on unique dates
for  i in goodIDs:
    glc = glims.loc[glims['GLAC_ID'] == i]
    if (len(glc.DATE.unique()) >= 4):
        greatIDs.append(i)
    
    #this appends to greatIDs glaciers with 4 or more dates
```

```python
print(len(greatIDs))
```

```python
#making a subplot to test for errors in data duplication 


dfEx = glims.loc[glims['GLAC_ID']==greatIDs[50]]
fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.scatter(dfEx['YEAR'], dfEx['DB_AREA'], color='green', marker='o')


#need to check 5 & 50 especially 


```

```python
#new data frame with all the good glaciers 

clean_ids = glims.loc[glims['GLAC_ID']==greatIDs[0]]

for i in np.arange(1,len(greatIDs)):
    dfEx = glims.loc[glims['GLAC_ID']==greatIDs[i]]
    clean_ids = clean_ids.append(dfEx)

clean_ids.head(10)
```

```python
clean_ids = clean_ids.drop(['Unnamed: 0'], axis=1)
```
