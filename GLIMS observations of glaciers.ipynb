```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
glims = pd.read_csv('glims_clean.csv')
```

```python
numGlacierObs = glims['GLAC_ID'].value_counts()
```

```python
#Let's try to get the "good IDs"

#To start, these are the glaciers with more than 3 observations
allIDs = glims.GLAC_ID.unique()
goodIDs = []
for i in allIDs:
    if numGlacierObs[i] > 3:
        goodIDs.append(i)
    
print(len(goodIDs))

#Now find glaciers with 4 or more obs on unique dates
for  i in goodIDs:
    glc = glims.loc[glims['GLAC_ID'] == i]
    uniqueDates = len(glc.DATE.unique())
    if uniqueDates < 4:
        goodIDs.remove(i)
    
print(len(goodIDs))
```
