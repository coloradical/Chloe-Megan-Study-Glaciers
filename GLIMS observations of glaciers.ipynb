```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
glims = pd.read_csv('glims_clean.csv')
```

```python
glims.head()
print(len(glims))
```

```python
to_drop = ['Randolph Glacier Inventory; Umbrella RC for merging the RGI into GLIMS']
glims = glims[~glims['GEOG_AREA'].isin(to_drop)]
```

```python
glims.head()
print(len(glims))
```

```python
numGlacierObs = glims['GLAC_ID'].value_counts()
```

```python
fig,ax = plt.subplots(1,1,figsize=(6,4))
ax.hist(numGlacierObs, bins = 14, edgecolor='white')
```

```python
#Let's try to get the "good IDs"

#To start, these are the glaciers with more than 3 observations
allIDs = glims.GLAC_ID.unique()
goodIDs = []
for i in allIDs:
    if numGlacierObs[i] > 3:
        goodIDs.append(i)
    
print(len(goodIDs))
```

```python
#Now find glaciers with 4 or more obs on unique dates
greatIDs = []
for  i in goodIDs:
    glc = glims.loc[glims['GLAC_ID'] == i]
    uniqueDates = len(glc.DATE.unique())
    if uniqueDates > 3:
        greatIDs.append(i)
    
print(len(greatIDs))
#how many are in the antartic peninsula? seems like a lot
```

```python
dfEx = glims.loc[glims['GLAC_ID'] ==greatIDs[4141]]
fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.scatter(dfEx['YEAR'],dfEx['DB_AREA'],color='green', marker='o')
```
